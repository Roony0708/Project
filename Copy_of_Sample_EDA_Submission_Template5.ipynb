## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# convert reviews type as integer type
pa_df = pa_df.astype({"Reviews": int})
````

````python
#drop $ sign from price column
pa_df['Price'] = pa_df['Price'].str.replace('$','')
````

````python
#convert price column type as float
pa_df = pa_df.astype({"Price": float})
````

````python
#in installs column we replace some sign and convert it into integer data type
pa_df['Installs'] = pa_df['Installs'].astype(int)
````

````python
#convert all apps size in mb data and drop some sign for coverting it into float data type
pa_df['Size'] = pa_df['Size'].apply(lambda x: str(x).replace('Varies with device', '20') if 'Varies with device' in str(x) else x)
pa_df['Size'] = pa_df['Size'].apply(lambda x: str(x).replace('M', '') if 'M' in str(x) else x)
pa_df['Size'] = pa_df['Size'].apply(lambda x: str(x).replace(',', '') if 'M' in str(x) else x)
pa_df['Size'] = pa_df['Size'].apply(lambda x: float(str(x).replace('k', '')) / 1000 if 'k' in str(x) else x)
pa_df['Size'] = pa_df['Size'].apply(lambda x: float(x))
````

````python
pa_df.info()
````

````python
#data based on rating vs price
price_df=pa_df[["App","Price","Type","Rating","Size"]]
price_df.head()
````

````python
#apps content rating
grouped = pa_df[['App','Content Rating']]
grouped
````

````python
#apps age group
age_grouped= grouped.rename(columns={'Content Rating': 'age_group'})
age_grouped
````

````python
#android version of apps
versions=pa_df["Android Ver"].value_counts().reset_index()
versions = versions.rename(columns={'index':'Android Ver','Android Ver': 'count'})
versions
````

````python
#total apps based on category
df=pa_df.groupby('Category')['App'].nunique().reset_index(name="Total Apps")
df=df.sort_values(by=['Total Apps'],ascending=False)
df
````
