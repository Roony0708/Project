## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
pa_df['Size'] = pa_df['Size'].map(lambda x: x.rstrip('M'))
pa_df['Size'] = pa_df['Size'].map(lambda x: str(round((float(x.rstrip('k'))/1024), 1)) if x[-1]=='k' else x)
pa_df['Size'] = pa_df['Size'].map(lambda x: 0 if x.startswith('Varies') else x)
````

````python
pa_df['Installs'] = pa_df['Installs'].map(lambda x: (x.rstrip('+')))
pa_df['Installs'] = pa_df['Installs'].map(lambda x: ''.join(x.split(','))) #since the figures are written in readable format which we will not be able to convert into numeric value otherwise
````

````python
pa_df['Installs'] = pd.to_numeric(pa_df['Installs'])
pa_df['Rating'] = pd.to_numeric(pa_df['Rating'])
pa_df['Reviews'] = pd.to_numeric(pa_df['Reviews'])
````

````python
#data based on rating vs price
price_df=pa_df[["App","Price","Type","Rating","Size"]]
price_df.head()
````

````python
#apps content rating
grouped = pa_df[['App','Content Rating']]
grouped
````

````python
#apps age group
age_grouped= grouped.rename(columns={'Content Rating': 'age_group'})
age_grouped
````

````python
#android version of apps
versions=pa_df["Android Ver"].value_counts().reset_index()
versions = versions.rename(columns={'index':'Android Ver','Android Ver': 'count'})
versions
````

````python
#total apps based on category
df=pa_df.groupby('Category')['App'].nunique().reset_index(name="Total Apps")
df=df.sort_values(by=['Total Apps'],ascending=False)
df
````
